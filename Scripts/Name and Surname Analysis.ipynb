```python
def get_surname_list():
    df = pd.read_csv("indian_surname.csv")
    df = df[['caste']]
    surname_list = df.values.tolist()
    cleaned_surname_list = []
    for state in surname_list:
        for word in state:
            word = word.lower()
            if len(word)>2:
                cleaned_surname_list.append(word)
    cleaned_surname_list.remove('more')
    cleaned_surname_list.remove('hole')
    cleaned_surname_list.remove('darling')
    cleaned_surname_list.remove('rai')
    cleaned_surname_list.remove('pai')
    cleaned_surname_list.remove('lad')
    cleaned_surname_list.remove('pal')
    cleaned_surname_list.remove('dev')
    cleaned_surname_list.remove('amin')
    cleaned_surname_list.remove('dang')
    cleaned_surname_list.remove('roy')
    cleaned_surname_list.remove('parate')
    cleaned_surname_list.remove('ingle')
    cleaned_surname_list.remove('lal')
    cleaned_surname_list.remove('ogra')
    cleaned_surname_list.remove('kar')
    cleaned_surname_list.remove('das')
    return cleaned_surname_list
```

```python
def get_demographics(para,city_list):
    demo_dict = {}
    for i in city_list:
        demo_dict[i] = 0
    para = para.lower()
    para_words = para.split(" ")
    for word in city_list:
        if word in para_words:
            demo_dict[word.lower()]=1
    sorted_demo_dict = {k: v for k, v in sorted(demo_dict.items(), key=lambda item: item[1],reverse=True)}
    return sorted_demo_dict
```

```python
def get_final_dict_city(period_list):
    dict_list = []
    cleaned_name_list = get_name_list()
    #cleaned_surname_list = get_surname_list()
    title_list = ['mr','dr','inspector']
    file_list = create_file_list(period_list)
    for file in file_list:
        para = get_para(file)
        #print(para)
        demo_dict = get_title_demographics(para,title_list)
        dict_list.append(demo_dict)
    result = dict(functools.reduce(operator.add, map(collections.Counter, dict_list)))
    sorted_city_dict = {k: v for k, v in sorted(result.items(), key=lambda item: item[1],reverse=True)}
    return sorted_city_dict
```
